Código que distribuye las cantidades totales por día, en cantidades por horario.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
pd.set_option('display.max_columns', None)

from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv('distribucion_dia_horario.csv', sep=';')
df.head()
```

```python
df.info()
```

```python
PE_Conditions = [
                    (df['Dia'] == 1),
                    (df['Dia'] == 2),
                    (df['Dia'] == 3),
                    (df['Dia'] == 4),
                    (df['Dia'] == 5),
                    (df['Dia'] == 6),
                    (df['Dia'] == 7)
]
PE_Categories = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo']
df['nombre_dia'] = np.select(PE_Conditions, PE_Categories)
print(df)
```

```python
df.Horas.value_counts(dropna=False)
```

```python
df1= pd.DataFrame(df.groupby(['nombre_dia', 'Horas'])['Cantidad'].first())
df1
```

```python
df2= df1.unstack(level=1)
```

```python
df2 = df2.droplevel(0, axis='columns')
```

```python
df2.reset_index()
```

```python
df2
```

```python
##Me traigo el segundo set de datos para unir.
#Paso siguiente, hacer la conversión 
```

```python
datp = pd.read_excel('Cantidad de público por día.xlsx')
datp
```

```python
datp['dia_semana'] = datp['Fecha'].dt.dayofweek
```

```python
datp
```

```python
PE_Conditions = [
                    (datp['dia_semana'] == 0),
                    (datp['dia_semana'] == 1),
                    (datp['dia_semana'] == 2),
                    (datp['dia_semana'] == 3),
                    (datp['dia_semana'] == 4),
                    (datp['dia_semana'] == 5),
                    (datp['dia_semana'] == 6)
]
PE_Categories = ['Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado', 'Domingo']
datp['nombre_dia'] = np.select(PE_Conditions, PE_Categories)
print(datp)
```

```python
print(datp.shape)
print(df2.shape)
data= datp.merge(df2, on='nombre_dia', how='left')
print(data.shape)
```

```python
data.head()
```

```python
data = data.rename(columns={ 6: '6h', 7: '7h', 8: '8h', 9: '9h', 10:'10h', 11: '11h', 12: '12h', 13:'13h', 14: '14h', 15: '15h', 16:'16h', 17: '17h', 18: '18h', 19:'19h', 20:'20h', 21:'21h', 22: '22h', 23: '23h'})
```

```python
data['suma']= data['6h'] + data['7h']+ data['8h']+ data['9h']+ data['10h'] + data['11h'] + data['12h'] + data['13h'] + data['14h'] + data['15h'] + data['16h'] + data['17h'] + data['18h'] + data['19h'] + data['20h'] + data['21h'] + data['22h']  + data['23h']
```

```python
data['unidad']= data['Cant. Público'] / data['suma']
```

```python
data.head()
```

```python
data['6c']= (data['unidad'] * data['6h']).astype(int)
data['7c']= (data['unidad'] * data['7h']).astype(int)
data['8c']= (data['unidad'] * data['8h']).astype(int)
data['9c']= (data['unidad'] * data['9h']).astype(int)
data['10c']= (data['unidad'] * data['10h']).astype(int)
data['11c']= (data['unidad'] * data['11h']).astype(int)
data['12c']= (data['unidad'] * data['12h']).astype(int)
data['13c']= (data['unidad'] * data['13h']).astype(int)
data['14c']= (data['unidad'] * data['14h']).astype(int)
data['15c']= (data['unidad'] * data['15h']).astype(int)
data['16c']= (data['unidad'] * data['16h']).astype(int)
data['17c']= (data['unidad'] * data['17h']).astype(int)
data['18c']= (data['unidad'] * data['18h']).astype(int)
data['19c']= (data['unidad'] * data['19h']).astype(int)
data['20c']= (data['unidad'] * data['20h']).astype(int)
data['21c']= (data['unidad'] * data['21h']).astype(int)
data['22c']= (data['unidad'] * data['22h']).astype(int)
data['23c']= (data['unidad'] * data['23h']).astype(int)
```

```python
data['sumados']= data['6c'] + data['7c']+ data['8c']+ data['9c']+ data['10c'] + data['11c'] + data['12c'] + data['13c'] + data['14c'] + data['15c'] + data['16c'] + data['17c'] + data['18c'] + data['19c'] + data['20c'] + data['21c'] + data['22c']  + data['23c']
```

```python
data.columns
```

```python
data = data[['Fecha', 'Shopping', 'Cant. Público', 'dia_semana', 'nombre_dia', '6c',
       '7c', '8c', '9c', '10c', '11c', '12c', '13c', '14c', '15c', '16c',
       '17c', '18c', '19c', '20c', '21c', '22c', '23c']]
```
